```python
from statistics import mean
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv("salary.csv", sep=',')
data.head()
```

```python
x = data['YearsExperience']
y = data['Salary']

x = x.values.reshape(len(x),1)
y = y.values.reshape(len(y),1)
```

```python
plt.scatter(x, y)
plt.show()
```

```python
def slope(x, y):
    m = (mean(x.flatten())*mean(y.flatten()) - mean(x.flatten()*y.flatten())) / ( mean(x.flatten())**2 -  mean(x.flatten()**2))
    b = mean(y.flatten()) - m * mean(x.flatten())
    return m, b
```

```python
m, b = slope(x, y)
```

```python
plt.scatter(x, y)
plt.plot(x , [m*_ + b for _ in x])
plt.show()
```
