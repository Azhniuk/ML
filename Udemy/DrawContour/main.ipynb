```python
import tkinter as tk
from tkinter import filedialog

import cv2 
from matplotlib import pyplot as plt
import numpy as np



class Example(tk.Frame):
    def __init__(self, parent):        
        frame = tk.Frame.__init__(self, parent)
        
        #button
        self.first_image_button = tk.Button(frame, text="Add Image", 
                                     relief="raised",
                                     command = lambda: self.selectImage())
        self.first_image_button.config(height = 4, width = 11 )
        self.first_image_button.place(relx = 0.08, rely = 0.08, anchor = 'center')
        self.first_image_button.state = True
        
        self.first_image_view = tk.Canvas()
        
        
    def selectImage(self):
        self.firstImagePath = filedialog.askopenfilename()
        #self.first_image = Image.open(self.first_image_path)
       
        self.image = cv2.imread(self.firstImagePath)
        self.image = cv2.resize(self.image, (500,500)) 
        self.findContour()
        
        
        
    def findContour (self):
        self.imageGray = cv2.cvtColor (self.image, cv2.COLOR_BGR2GRAY)
        estimatedThreshold, imageThreshold = cv2.threshold(self.imageGray,140,255,cv2.THRESH_BINARY)
        contours, hierarchy = cv2.findContours(imageThreshold,cv2.RETR_CCOMP,cv2.CHAIN_APPROX_SIMPLE)
        
        imageCopy = np.copy(self.image)
        
        for i, c in enumerate(contours):
            areaContour=cv2.contourArea(c)
            if areaContour<2000 or 100000<areaContour:
                continue
            cv2.drawContours(imageCopy,contours,i,(255,10,255),4)
        cv2.imshow('Image', imageCopy)
        cv2.waitKey(0)
        cv2.destroyAllWindows()
        



if __name__ == "__main__":    
    #create a window and set parameters of a window
    root = tk.Tk()
    width = 800
    height = 600
    
    #center it    
    screen_width = root.winfo_screenwidth()
    screen_height = root.winfo_screenheight()
    center_x = int(screen_width / 2 - width / 2)
    center_y = int(screen_height / 2 - height / 2)
        
    root.geometry(f'{width}x{height}+{center_x}+{center_y}')
    root.title('Find contour')
        
    Example(root).pack( fill = 'both' , expand = True)
    
    root.mainloop()
  
        
```
