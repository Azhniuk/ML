```python
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
iris = load_iris()
iris = iris["data"]
```

```python
len(iris)
```

```python
x = [_[0] for _ in iris]
y = [_[1] for _ in iris ]
plt.scatter(x ,y )
plt.xlabel('length')
plt.ylabel('width')
plt.show()
```

# З вхідними даними х у.

```python
data = list(zip(x, y))

inertias = []

for i in range(1,15):
    kmeans = KMeans(n_clusters=i)
    kmeans.fit(data)
    inertias.append(kmeans.inertia_)

plt.plot(range(1,15), inertias, marker='o')
plt.title('Elbow method')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.show()

#по приколу, нам це не потрібно бо 3 види
```

```python
kmeans = KMeans(n_clusters=3)
kmeans.fit(data)

plt.scatter(x, y, c=kmeans.labels_)
plt.show()
```
