```python
import matplotlib.pyplot as plt

x = [4,  5,  10, 4,  3,  11, 14 , 6, 10, 12]
y = [12, 14, 24, 17, 10, 25, 24, 12, 21, 21]

plt.scatter(x, y)
plt.show()
```

# With library

```python
from sklearn.cluster import KMeans

data = list(zip(x, y))
#робить list of tuples об'єднує х з у

inertias = []

for i in range(1,11):
    kmeans = KMeans(n_clusters=i)
    kmeans.fit(data)
    inertias.append(kmeans.inertia_)

plt.plot(range(1,11), inertias, marker='o')
plt.title('Elbow method')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.show()
```

```python
kmeans = KMeans(n_clusters=2)
kmeans.fit(data)

plt.scatter(x, y, c=kmeans.labels_)
plt.show()
```

# Without library
*припустимо: нам дано, що має бути два класи

```python
data = list(zip(x, y))

C1 = data[0]
C2 = data[1]
#майже randomly selected centroids

x_c1, x_c2, y_c1, y_c2 =  0, 0, 0, 0
```

```python
def group(data, one= 0, two = 0):
    groups = []
    for i in data:
        len1 = ((i[0]-C1[0])**2 + (i[1]-C1[1])**2)**0.5
        len2 = ((i[0]-C2[0])**2 + (i[1]-C2[1])**2)**0.5
        if len1 < len2:
            groups.append(1)
            one += 1
        else:
            groups.append(2)
            two += 1
    return groups, one, two
```

```python
groups, one, two = group(data)
```

```python
# recalculate the centroid
for i in groups:
    if i == 1:
        x_c1 += data[i][0]
        y_c1 += data[i][1]
        
    else:
        x_c2 += data[i][0]
        y_c2 += data[i][1]
        
C1 = tuple(map(int, [x_c1/one, y_c1/one]))
C2 = tuple(map(int, [x_c2/two, y_c2/two]))
```

```python
plt.scatter(x, y)
plt.scatter(C1[0], C1[1])
plt.scatter(C2[0], C2[1])
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
