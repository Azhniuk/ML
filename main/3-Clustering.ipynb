```python
names = ['id','title','description','link','image_url','availability','availability_date','price','brand','condition','mpn','gtin','product_type','google_product_category','custom_label_0','custom_label_1','custom_label_2']
```

```python
df = pd.read_csv('/Users/Sofiia/Documents/GitHub/txt.csv', sep='\t', names=names, skiprows=1)
df['price_new'] = df['price'].apply(lambda x: float(x[:-4]))
```

```python
arr = np.array([])
```

```python
for _ in df['price_new']:
    arr = np.append(arr, _)
```

```python
import matplotlib.pyplot as plt
plt.scatter(arr[0:140], arr[0:140])
plt.show()
#найбільше від 0 до 2000
```

-

```python
from sklearn.cluster import KMeans

data = list(zip(arr[0:140], arr[0:140]))
#робить list of tuples об'єднує х з у

inertias = []

for i in range(1,11):
    kmeans = KMeans(n_clusters=i)
    kmeans.fit(data)
    inertias.append(kmeans.inertia_)

plt.plot(range(1,11), inertias, marker='o')
plt.title('Elbow method')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.show()
```

```python
kmeans = KMeans(n_clusters=3)
kmeans.fit(data)

plt.scatter(arr[0:140], arr[0:140], c=kmeans.labels_)
plt.show()
```
